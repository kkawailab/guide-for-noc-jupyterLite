### 追加例: 2つの周波数を重ねたビート

```python
import math

f1, f2 = 2.0, 2.2  # Hz
samples = []
for i in range(20):
    t = i / 10.0
    y = math.sin(2 * math.pi * f1 * t) + math.sin(2 * math.pi * f2 * t)
    samples.append(round(y, 3))
print("合成波サンプル:", samples)
```
