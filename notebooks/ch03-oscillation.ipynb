- Simple harmonic motion arises from restoring forces proportional to displacement (Hooke).
- Phase, amplitude, and angular frequency describe oscillators compactly.
- Damping reduces amplitude over time; under/over/critical damping define regimes.
- Combine sine waves or use polar coordinates for rotational motion.

Try: Set damping to 0 for perpetual oscillation or >0.1 for overdamped motion.
