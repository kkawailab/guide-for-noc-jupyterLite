# Chapter 05 — 物理: 簡易エンジン風ステップ

- 簡易ワールドで複数ボディに重力と床衝突を適用。
- 実際のエンジンでは衝突判定や拘束をライブラリが肩代わりする。

試してみよう: コード内の定数を調整して挙動の変化を確認。

```javascript
let bodies = [];function setup() {  createCanvas(360, 320);  for (let i = 0; i < 3; i++) {    bodies.push({ pos: createVector(60 + i*80, 40), vel: createVector(0,0) });  }}function draw() {  background(250);  for (const b of bodies) {    b.vel.y += 0.4;    b.pos.add(b.vel);    if (b.pos.y > height-12) { b.pos.y = height-12; b.vel.y *= -0.6; }    fill(60);    circle(b.pos.x, b.pos.y, 20);  }}
```
