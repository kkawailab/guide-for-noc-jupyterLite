# Chapter 02 — 力: 重力と抗力

- 重力を一定の加速度、抗力を速度に比例する減速として加える。
- 抗力を強くすると終端速度が低くなる。

試してみよう: コード内の定数を調整して挙動の変化を確認。

```javascript
let y, v;function setup() {  createCanvas(320, 320);  y = 20;  v = 0;}function draw() {  background(245);  const g = 0.4;  const drag = 0.02 * v * v * Math.sign(v);  v += g - drag;  y += v;  if (y > height-12) { y = height-12; v *= -0.6; }  fill(50, 120, 200);  circle(width/2, y, 24);  text(`v=${v.toFixed(2)}`, 10, 20);}
```
