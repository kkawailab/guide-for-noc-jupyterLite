# Chapter 08 — フラクタル: 再帰的な樹

- 再帰で枝を分岐させて自己相似の樹を描く。
- 角度と縮小率で形が大きく変わる。

試してみよう: コード内の定数を調整して挙動の変化を確認。

```javascript
function setup() {  createCanvas(360, 360);  background(245);  stroke(40);  translate(width/2, height);  branch(90, 0);}function branch(len, angle) {  const theta = radians(25);  const newLen = len * 0.7;  const dx = newLen * sin(angle);  const dy = newLen * cos(angle);  line(0, 0, dx, -dy);  translate(dx, -dy);  if (newLen > 10) {    push(); branch(newLen, angle + theta); pop();    push(); branch(newLen, angle - theta); pop();  }}
```
