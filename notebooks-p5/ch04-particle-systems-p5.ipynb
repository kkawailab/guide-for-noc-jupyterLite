# Chapter 04 — パーティクル: 噴水エミッタ

- 原点から寿命つき粒子を連続生成し、重力で放物線を描かせる。
- 速度分布や寿命で見た目が変わる。

試してみよう: コード内の定数を調整して挙動の変化を確認。

```javascript
// パーティクルの状態と生成/更新ヘルパー
let particles = [];

function makeParticle() {
  return {
    pos: createVector(width / 2, height - 10),
    vel: createVector(random(-1, 1), random(-4, -7)),
    life: 80,
  };
}

function updateParticle(p) {
  p.vel.y += 0.12;
  p.pos.add(p.vel);
  p.life -= 1;
}

function drawParticle(p) {
  const alpha = map(p.life, 0, 80, 0, 255);
  noStroke();
  fill(80, 120, 255, alpha);
  circle(p.pos.x, p.pos.y, 8);
}
```

```javascript
// 毎フレームの生成・更新・描画
function setup() {
  createCanvas(420, 300);
}

function draw() {
  background(245);
  particles.push(makeParticle());
  for (const p of particles) updateParticle(p);
  particles = particles.filter((p) => p.life > 0);
  for (const p of particles) drawParticle(p);
  text(`count: ${particles.length}`, 10, 20);
}
```
