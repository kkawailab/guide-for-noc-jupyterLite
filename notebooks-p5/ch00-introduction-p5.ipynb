# Chapter 00 — 序章: ランダムウォーク

- ランダムな歩みを線で可視化して、軌跡から性質を観察する。
- ステップ幅やバイアスを変えると拡散の形が変わる。

試してみよう: コード内の定数を調整して挙動の変化を確認。

```javascript
let walker;function setup() {  createCanvas(420, 420);  background(245);  walker = createVector(width/2, height/2);}function draw() {  stroke(20, 80);  strokeWeight(2);  point(walker.x, walker.y);  const step = p5.Vector.random2D().mult(8);  walker.add(step);  walker.x = constrain(walker.x, 0, width);  walker.y = constrain(walker.y, 0, height);}
```
