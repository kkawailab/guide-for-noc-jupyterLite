# Chapter 01 — ベクトル: 速度制限でターゲットに向かう

- 位置・速度をベクトルで扱い、速度を上限で制御する。
- 目標へ向かうdesiredベクトルと現在速度の差をステアリングとする。

試してみよう: コード内の定数を調整して挙動の変化を確認。

```javascript
let pos, vel, target;function setup() {  createCanvas(420, 240);  pos = createVector(40, height/2);  vel = createVector(0, 0);  target = createVector(width-40, height/2);}function draw() {  background(240);  const desired = p5.Vector.sub(target, pos).setMag(3);  const steer = p5.Vector.sub(desired, vel).limit(0.2);  vel.add(steer).limit(3);  pos.add(vel);  fill(30);  circle(pos.x, pos.y, 16);  fill(200,0,0);  circle(target.x, target.y, 10);}
```
