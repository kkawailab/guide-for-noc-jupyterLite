# Chapter 06 — 自律エージェント: wander

- 進行方向付近の円上にランダム目標を置き、滑らかに徘徊させる。
- ステアリング力と速度を制限して自然な動きに。

試してみよう: コード内の定数を調整して挙動の変化を確認。

```javascript
// エージェントの状態とキャンバス初期化
let agent;

function setup() {
  createCanvas(420, 260);
  agent = { pos: createVector(width / 2, height / 2), vel: createVector(2, 0), theta: 0 };
}
```

```javascript
// 徘徊ステアリングで位置を更新
function draw() {
  background(245);
  wander(agent);
  agent.pos.add(agent.vel);
  agent.pos.x = (agent.pos.x + width) % width;
  agent.pos.y = (agent.pos.y + height) % height;
  fill(30);
  circle(agent.pos.x, agent.pos.y, 14);
}
```

```javascript
// ランダム方向を混ぜるwander関数
function wander(a) {
  a.theta += random(-0.3, 0.3);
  const circleDir = createVector(cos(a.theta), sin(a.theta));
  const desired = p5.Vector.add(a.vel.copy().normalize(), circleDir).setMag(2.5);
  const steer = p5.Vector.sub(desired, a.vel).limit(0.15);
  a.vel.add(steer).limit(3);
}
```
