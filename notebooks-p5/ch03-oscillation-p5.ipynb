# Chapter 03 — 振動: バネの単振動

- フックの法則で復元力、速度に比例した減衰を入れて単振動を描く。
- 減衰を0にすると永続振動になる。

試してみよう: コード内の定数を調整して挙動の変化を確認。

```javascript
// バネのパラメータと初期状態
let x = 180;
let v = 0;
const k = 0.05; // バネ定数
const damping = 0.02;
const anchor = 200;

function setup() {
  createCanvas(400, 180);
}
```

```javascript
// フックの法則と減衰を適用して描画
function draw() {
  background(250);
  const stretch = x - anchor;
  const force = -k * stretch - damping * v;
  v += force;
  x += v;
  line(anchor, 40, x, 40);
  fill(40);
  circle(x, 40, 22);
  text(`x=${x.toFixed(1)}`, 10, 20);
}
```
